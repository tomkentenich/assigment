**Importing the libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import cross_val_score 
```

**Importing the dataset**

```python
df=pd.read_csv('Social_Network_Ads.csv')
df
```

```python
#df=df.drop(columns='User ID')
label_encoder =LabelEncoder()
df['Gender']=label_encoder.fit_transform(df['Gender'])
```

```python
df
```

**Splitting the dataset into the Training set and Test set**

```python
X=df[['Gender','Age','EstimatedSalary']]
y=df['Purchased']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

**Feature Scaling**

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

**Fitting K-NN to the Training set**

```python
def knn_compare(data,i,X):
  y=df['Purchased'].values
  clf=KNeighborsClassifier(i)
  clf.fit(X,y)
  plot_decision_regions(X,y,clf=clf,legend=2)
  print("K : ", i, " Train Accuracy : ", clf.score(X,y), " Val Accuracy : ", np.mean(cross_val_score(clf, X, y, cv=10)))
  plt.xlabel('X') 
  plt.ylabel('Y') 
  plt.title('Knn with K='+ str(i))
  plt.show() 
```

```python
from sklearn.metrics import plot_roc_curve,roc_curve,roc_auc_score,confusion_matrix,f1_score,classification_report
for i in range(1,100,10):
  X=df[['Gender','EstimatedSalary']].values
  knn_compare(df,i,X)
#print('K value = ',i,'score = ',roc_auc_score(y_test,knn.predict_proba(X_test)[:,1]))
```
