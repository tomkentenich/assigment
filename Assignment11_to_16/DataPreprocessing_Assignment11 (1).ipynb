**Step 1: Importing the libraries**

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

**Step 2: Importing dataset**

```python
df=pd.read_csv('Data.csv')
```

```python
df
```

**Step 3: Handling the missing data**

```python
df.info()
```

```python
df.loc[df['Age'].isnull()]
```

```python
age=df.loc[(df['Country']=='Spain') & (df['Age'].notnull()),'Age'].mean()
df['Age']=df['Age'].fillna(int(age))
```

```python
df.loc[df['Salary'].isnull()]
```

```python
salary=df.loc[(df['Country']=='Germany')&(df['Salary'].notnull()),'Salary'].mean()
df['Salary']=df['Salary'].fillna(salary)
```

```python
df
```

**Step 4: Encoding categorical data**

```python
label_encoder = preprocessing.LabelEncoder()
df['Purchased']=label_encoder.fit_transform(df['Purchased'])
```

**Step 5: Creating a dummy variable**

```python
dummies=pd.get_dummies(df.Country)
dummies
```

```python
final=pd.concat([df,dummies],axis='columns')
final
```

```python
#final.drop(columns='Country')
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
X=final[["Age", "Salary", "Germany", "Spain"]]
y=final["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

**Step 7: Feature Scaling**

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```
