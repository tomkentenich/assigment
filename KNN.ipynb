##Spliting the data into train and split



```python
X_train=df.iloc[:,:-1]
y_train=df.iloc[:,-1]
X_test=df.iloc[:,:-1]
y_test=df.iloc[:,-1]
```

##Scaling the data

```python
scaler=StandardScaler()
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

##Fitiing the KNN

```python
for k in range(1,10):
  X=X_train,y_train
  knn=KNeighborsClassifier(k)
  knn.fit(X_train,y_train)
  print("K : ", k, " Train Accuracy : ", knn.score(X_train,y_train), " cross Value Accuracy : ", np.mean(cross_val_score(knn,X_train,y_train, cv=2)))
```

```python
for k in range(2,4):
  X=X_train,y_train
  knn=KNeighborsClassifier(1)
  knn.fit(X_train,y_train)
  print("K : ", 1, " Train Accuracy : ", knn.score(X_train,y_train), " cross Value Accuracy : ",k," is ", np.mean(cross_val_score(knn,X_train,y_train, cv=k)))
```

```python
X=X_train,y_train
knn=KNeighborsClassifier(1)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```
